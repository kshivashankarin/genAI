```python
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Arun", "Priya", "Kavin"]
})

# Second DataFrame
df2 = pd.DataFrame({
    "Name": ["Arun", "Priya", "Kavin"],
    "ID": [1, 2, 3],
    "Marks": [85, 90, 78]
})

# Merge using common column "ID"
merged_df = pd.merge(df1, df2, on="ID")

display(merged_df)
```
